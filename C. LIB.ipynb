```python
def get_character(episode):
    len_ch = 0
    LINES = pd.DataFrame(open(episode, 'r', encoding='utf-8-sig').readlines(), columns=['line_str'])
    for token in LINES['line_str']:
        if isinstance(token, str):  # Check if the element is a string
            len_ch += len(token)
        elif isinstance(token, float):  # Check if the element is a float
            len_ch += len(str(token))  # Convert float to string and get its length
    return len_ch
```
